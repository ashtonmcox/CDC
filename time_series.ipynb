```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_pickle("final_features.pkl")

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['fare'] + 2000, marker='o', linestyle='-', color='b', label='Value')
plt.title('Time Series Line Plot')
plt.xlabel('Year')
plt.ylabel('fare')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
