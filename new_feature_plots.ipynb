```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_pickle("final_features.pkl")

#  scatter plot
sns.scatterplot(data=df, x='Year', y='passengers', palette='deep')

# Add titles labels
plt.title('Year vs. Passengers')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.show()
```

```python
yearly_passengers = df.groupby('Year')['passengers'].sum().reset_index()

# Create histogram
sns.barplot(data=yearly_passengers, x='Year', y='passengers', palette='deep')

# Add titles and labels
plt.title('Total Passengers per Year')
plt.xlabel('Year from Start')
plt.ylabel('Total Passengers')

# Rotate x-ticks
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=df, x='fare', y='fare_low', palette='deep')
sns.regplot(data=df, x='fare', y='fare_low', scatter_kws={'s': 20}, line_kws={'color': 'red'})


# Add titles labels
plt.title('Fare vs. Fare Low')
plt.xlabel('Fare')
plt.ylabel('Fare Low')

# Show plot
plt.show()
```

```python
sns.scatterplot(data=df, x='Year', y='fare', palette='deep')

# Add titles and labels
plt.title('Year vs. Fare')
plt.xlabel('Year')
plt.ylabel('Fare')

# Show plot
plt.show()
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15, 12))

# Scatter plot between nsmiles and passengers
axs[0, 0].scatter(df['nsmiles'], df['passengers'], alpha=0.6, color='b')
axs[0, 0].set_title('nsmiles vs passengers')
axs[0, 0].set_xlabel('nsmiles')
axs[0, 0].set_ylabel('passengers')

# Line plot for fare over Year
axs[0, 1].plot(df['Year'], df['fare'], color='r')
axs[0, 1].set_title('Year vs fare')
axs[0, 1].set_xlabel('Year')
axs[0, 1].set_ylabel('fare')

# Histogram of city_distance
axs[0, 2].hist(df['city_distance'], bins=20, color='g')
axs[0, 2].set_title('city_distance distribution')
axs[0, 2].set_xlabel('city_distance')

# Boxplot of fare and fare_low
axs[1, 0].boxplot([df['fare'], df['fare_low']], labels=['fare', 'fare_low'])
axs[1, 0].set_title('Boxplot of fare and fare_low')

# Scatter plot between city1_lati and city2_lati
axs[1, 1].scatter(df['city1_lati'], df['city2_lati'], alpha=0.6, color='purple')
axs[1, 1].set_title('city1_lati vs city2_lati')
axs[1, 1].set_xlabel('city1_lati')
axs[1, 1].set_ylabel('city2_lati')

# Bar plot for city1_region_Midwest and city1_region_West
axs[1, 2].bar(['Midwest', 'West'], [df['city1_region_Midwest'].sum(), df['city1_region_West'].sum()], color='orange')
axs[1, 2].set_title('city1_region_Midwest vs city1_region_West')

# Scatter plot for fare_passenger_interaction vs passengers_per_mile
axs[2, 0].scatter(df['fare_passenger_interaction'], df['passengers_per_mile'], alpha=0.6, color='brown')
axs[2, 0].set_title('fare_passenger_interaction vs passengers_per_mile')
axs[2, 0].set_xlabel('fare_passenger_interaction')
axs[2, 0].set_ylabel('passengers_per_mile')

# Histogram for passenger_market_demand
axs[2, 2].hist(df['passenger_market_demand'], bins=20, color='red')
axs[2, 2].set_title('passenger_market_demand distribution')

plt.tight_layout()
plt.show()
```
