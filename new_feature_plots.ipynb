```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python


df = pd.read_pickle("final_features.pkl")

# Create scatter plot
sns.scatterplot(data=df, x='Year', y='passengers', palette='deep')


# Add titles and labels
plt.title('Year vs. Passengers')
plt.xlabel('Year')
plt.ylabel('Passengers')

# Show plot
plt.show()
```

```python
# Load the data
df = pd.read_pickle("final_features.pkl")

# Group by Year and calculate the total number of passengers per year
yearly_passengers = df.groupby('Year')['passengers'].sum().reset_index()

# Create histogram
sns.barplot(data=yearly_passengers, x='Year', y='passengers', palette='deep')

# Add titles and labels
plt.title('Total Passengers per Year')
plt.xlabel('Year' + 1993)
plt.ylabel('Total Passengers')

# Rotate x-ticks for better readability if needed
plt.xticks(rotation=45)

# Show plot
plt.show()
```
