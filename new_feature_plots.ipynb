```python
# Load the data
df = pd.read_pickle("final_features.pkl")

# Group by Year and calculate the total number of passengers per year
yearly_passengers = df.groupby('Year')['passengers'].sum().reset_index()

# Create histogram
sns.barplot(data=yearly_passengers, x='Year', y='passengers', palette='deep')

# Add titles and labels
plt.title('Total Passengers per Year')
plt.xlabel('Year' + 1993)
plt.ylabel('Total Passengers')

# Rotate x-ticks for better readability if needed
plt.xticks(rotation=45)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data creation for illustration (normally, we'd use real data)
features = ['Year', 'citymarketid_1', 'citymarketid_2', 'nsmiles', 'passengers',
       'fare', 'large_ms', 'fare_lg', 'lf_ms', 'fare_low', 'city1_lati',
       'city1_long', 'city2_lati', 'city2_long', 'airportid_1_encoded',
       'airportid_2_encoded', 'fare_passenger_interaction', 'city_distance',
       'fare_diff_lg_low', 'fare_low_lf_ms_interaction',
       'citymarketid_1_encoded', 'citymarketid_2_encoded', 'large_ms_std',
       'lf_ms_std', 'passengers_per_mile', 'passenger_market_demand',
       'miles/passenger', 'city1_popularity', 'city2_popularity',
       'nmiles/popularity1', 'city1_region_Midwest', 'city1_region_Northeast',
       'city1_region_Other', 'city1_region_South', 'city1_region_West',
       'city2_region_Midwest', 'city2_region_Northeast', 'city2_region_Other',
       'city2_region_South', 'city2_region_West']

# Create a fake DataFrame with random values for these features for demonstration
np.random.seed(42)  # For reproducible results
data = pd.DataFrame(np.random.rand(100, len(features)), columns=features)

# Plotting some of the main features for visualization
fig, axs = plt.subplots(3, 3, figsize=(15, 12))

# Scatter plot between 'nsmiles' and 'passengers'
axs[0, 0].scatter(data['nsmiles'], data['passengers'], alpha=0.6, color='b')
axs[0, 0].set_title('nsmiles vs passengers')
axs[0, 0].set_xlabel('nsmiles')
axs[0, 0].set_ylabel('passengers')

# Line plot for 'fare' over 'Year'
axs[0, 1].plot(data['Year'], data['fare'], color='r')
axs[0, 1].set_title('Year vs fare')
axs[0, 1].set_xlabel('Year')
axs[0, 1].set_ylabel('fare')

# Histogram of 'city_distance'
axs[0, 2].hist(data['city_distance'], bins=20, color='g')
axs[0, 2].set_title('city_distance distribution')
axs[0, 2].set_xlabel('city_distance')

# Boxplot of 'fare' and 'fare_low'
axs[1, 0].boxplot([data['fare'], data['fare_low']], labels=['fare', 'fare_low'])
axs[1, 0].set_title('Boxplot of fare and fare_low')

# Scatter plot between 'city1_lati' and 'city2_lati'
axs[1, 1].scatter(data['city1_lati'], data['city2_lati'], alpha=0.6, color='purple')
axs[1, 1].set_title('city1_lati vs city2_lati')
axs[1, 1].set_xlabel('city1_lati')
axs[1, 1].set_ylabel('city2_lati')

# Bar plot for 'city1_region_Midwest' and 'city1_region_West'
axs[1, 2].bar(['Midwest', 'West'], [data['city1_region_Midwest'].sum(), data['city1_region_West'].sum()], color='orange')
axs[1, 2].set_title('city1_region_Midwest vs city1_region_West')

# Scatter plot for 'fare_passenger_interaction' vs 'passengers_per_mile'
axs[2, 0].scatter(data['fare_passenger_interaction'], data['passengers_per_mile'], alpha=0.6, color='brown')
axs[2, 0].set_title('fare_passenger_interaction vs passengers_per_mile')
axs[2, 0].set_xlabel('fare_passenger_interaction')
axs[2, 0].set_ylabel('passengers_per_mile')

# Line plot of 'large_ms' and 'lf_ms' over index
axs[2, 1].plot(data.index, data['large_ms'], label='large_ms', color='green')
axs[2, 1].plot(data.index, data['lf_ms'], label='lf_ms', color='blue')
axs[2, 1].set_title('large_ms vs lf_ms')
axs[2, 1].legend()

# Histogram for 'passenger_market_demand'
axs[2, 2].hist(data['passenger_market_demand'], bins=20, color='red')
axs[2, 2].set_title('passenger_market_demand distribution')

plt.tight_layout()
plt.show()
```
